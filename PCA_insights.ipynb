```python
import pandas as pd 
import texthero as hero
```

```python
dev_df_h = pd.read_csv('drive/MyDrive/pubhealth/data/clean/exp1/clean_dev.tsv',sep='\t')
train_df_h = pd.read_csv('drive/MyDrive/pubhealth/data/clean/exp1/clean_train.tsv',sep='\t')
```

```python
dev_df_h.head()
```

```python
#PCA for all columns  
```

```python
dev_df_h["pca_exp"] = (
             dev_df_h["explanation"]
            .pipe(hero.tfidf) # term frequency and inverse document frequency 
            .pipe(hero.pca)
)
```

```python
hero.scatterplot(
    dev_df_h, 
    col='pca_exp', 
    color='label', 
    title="PCA Explanation"
)
```

```python
dev_df_h["pca_main_text"] = (
             dev_df_h["main_text"]
            .pipe(hero.tfidf) # term frequency and inverse document frequency 
            .pipe(hero.pca)
)
```

```python
hero.scatterplot(
    dev_df_h, 
    col='pca_main_text', 
    color='label', 
    title="PCA Main text"
)
```

**Note**: cluster under/right to y = 3x - 0.3 

```python
dev_df_h["pca_claim"] = (
             dev_df_h["claim"]
            .pipe(hero.tfidf) # term frequency and inverse document frequency 
            .pipe(hero.pca)
)
```

```python
hero.scatterplot(
    dev_df_h, 
    col='pca_claim', 
    color='label', 
    title="PCA Claim"
)
```

 **Notes:**
 


*   Except for **subject** column, the rest of the columns are taken from **dev** set **instead of train set** due to lack of **RAM** (even with GPU google colab).
*   **Subject** in training set shows clearly that there are subjects that highly influence the Claim outcome/label when it comes to False and Unproven ones. (see Subject PCA plot)
*   There are noticeable **outliers** or extreme cases in all the PCAs.
*   A great region of data points are **< 0** in terms of PCA when it comes to **TF-IDF**.

**Decision**: 
Keep 'subject'column.


```python
train_df_h["pca_subjects"] = (
             train_df_h["subjects"]
            .pipe(hero.tfidf) # term frequency and inverse document frequency 
            .pipe(hero.pca)
)
```

```python
hero.scatterplot(
    train_df_h, 
    col='pca_subjects', 
    color='label', 
    title="PCA Subjects"
)
```

When It comes to **True** claims there is no great PCA component significance according to **subjects** but for other labls particularly **False and unproven** claims the tendency is clear.
